```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

lucasmean = pd.read_csv('csvs/mean_1M.csv', header = None).values
lucasstd = pd.read_csv('csvs/stdev_1M.csv', header = None).values
charliemean = pd.read_csv('csvs/ave_n_1M.csv').iloc[:, 1:].values
charliestd = pd.read_csv('csvs/stdev_n_1M.csv').iloc[:, 1:].values
```

```python
#plots a rectangular color plot
def colorplot(localz, title):
    plt.figure(figsize = (10,5))
    plt.imshow(localz, origin='upper')
    plt.xlabel('RA')
    plt.ylabel('DEC')
    plt.title(title)
    plt.colorbar()
    plt.show()
```

```python
meanpull = (2*(lucasmean-charliemean)/(lucasmean+charliemean))[:105]
stdpull = (2*(lucasstd-charliestd)/(lucasstd+charliestd))[:105]

colorplot(meanpull, 'mean pull')
colorplot(stdpull, 'std pull')

meanpulldf = pd.DataFrame(meanpull)
stdpulldf = pd.DataFrame(stdpull)

meanpulldf.to_csv('csvs/mean_pull.csv')
stdpulldf.to_csv('csvs/std_pull.csv')
```
